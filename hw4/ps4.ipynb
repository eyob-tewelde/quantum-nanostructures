```python
qw = np.array([25, 29, 33, 38, 42, 48, 53, 56, 60, 65, 69, 73, 75, 82, 84, 86, 89, 94, 98, 105, 111, 118, 129, 141])
eg_values = np.array([
    0.37882, 0.37195, 0.36743, 0.36366, 0.36157, 0.35935, 0.35804, 0.35742, 
    0.35674, 0.35605, 0.35561, 0.35523, 0.35507, 0.35454, 0.35443, 0.35436, 
    0.35427, 0.35414, 0.35412, 0.35408, 0.35407, 0.35407, 0.35405, 0.35405
])
```

```python
esc = 15.15
eo = 8.854188 * 10 ** -12
e = 1.60219 * 10 ** -19
h = 6.62618 * 10 ** -34
me = 9.10953 * 10 ** -31


def reduced_mass(bg, qwt):
    return (h ** 2) / (8 * (bg - 0.35) * qwt ** 2)

def bohr_rad(rm):
    return (esc * eo * h ** 2) / (np.pi * rm * e ** 2)
```

```python
reduced_m = reduced_mass(eg_values, qw / (10 ** 9))
br = bohr_rad(reduced_m)
```

```python
plt.plot(qw, br)
plt.gca().spines[["top", "right"]].set_visible(False)
plt.title(r"$a_b$ vs QWell Thickness", fontsize=13)
plt.xlabel("QWell Thickness (nm)")
plt.ylabel(r"$\text{a}_b$ (Ã…)", fontsize=10)
# plt.savefig("q1_ps3.png")
plt.show()
```
