```python
import matplotlib.pyplot as plt
import numpy as np

k = 8.6173303 * (10 ** -5)
```

```python
def fermi_dirac(e, t):
    num = e - 5.49
    denom = k * t
    exp = np.exp(num/denom)
    return 1 / (exp + 1)
```

```python
temp = np.array([10, 500, 2000, 8000])
energy = np.arange(2, 9, 0.1)

probability = []
for t in temp:
    probability.append(fermi_dirac(energy, t))
```

```python
for i, p in enumerate(probability):
    plt.plot(p, energy, label=f"{temp[i]} K")
    plt.legend()
    plt.gca().spines[["top", "right"]].set_visible(False)
    plt.xlabel('f (E)')
    plt.ylabel("Energy (eV)")
    plt.title("f(E) vs Energy")

plt.savefig("quantum_ps1_q5.png")
```
