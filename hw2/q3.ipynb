```python
import numpy as np
import matplotlib.pyplot as plt

e = np.array([.633,.645,.661,.673,.682,.692,.701,.706,.715,.725,.731,.740,.745,.754,.76,.767,.777])
a = np.array([0,.25,1,2.25,4,6.25,9,12.25,16,20.25,25,30.25,36,42.25,49,56.25,64])

plt.scatter(e,a ** 2)
plt.gca().spines[["top", "right"]].set_visible(False)
plt.xlabel("E (eV)")
plt.ylabel("a^2 (cm^-2)")
```
