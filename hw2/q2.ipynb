```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
e333 = [1.035,1.05, 1.065, 1.1, 1.12, 1.145, 1.17,1.18, 1.2, 1.22,1.24,1.25]
a333_sqrt = np.array([0,0.6, 1, 1.9, 2.3, 3, 4, 4.5,6, 7,8,9])
a333 = a333_sqrt ** 4

# plt.plot(e333[0:7], a333[0:7], color="blue")
# plt.plot(e333[6:], a333[6:], color="blue")
# plt.plot(e333, a333)
plt.scatter(e333, a333, color="red")
plt.gca().spines[["top", "right"]].set_visible(False)
plt.title("a vs E at T = 333 K")
plt.ylabel("a (cm^-1)")
plt.xlabel("E (eV)")
plt.savefig("ps2_q2a_333.png")
```

```python
e195 = [1.06, 1.1, 1.15, 1.16, 1.18, 1.19, 1.2,1.21, 1.22,1.23, 1.235,1.25]
a195_sqrt = np.array([0, 0.3, 1, 1.1, 1.8, 2, 2.5, 3,3.5,4, 4.2,5]) 
a195 = a195_sqrt ** 4

plt.scatter(e195,a195, color="red")
# plt.plot(e195, a195)
plt.gca().spines[["top", "right"]].set_visible(False)
plt.title("a vs E at T = 195 K")
plt.ylabel("a (cm^-1)")
plt.xlabel("E (eV)")
plt.savefig("ps2_q2a_195.png")
```

```python
temp = np.array([195, 249, 290, 333])
be = np.array([1.1, 1.085, 1.071, 1.053])

def empr(temp):
    return 1.17 + ((1.059 * 10 ** -5) * temp) + ((-6.05 * 10 ** -7) * (temp ** 2))

plt.scatter(temp, be, label="Data")
plt.plot(temp, empr(temp), color="orange", label="Emperical fit")
plt.gca().spines[["top", "right"]].set_visible(False)
plt.title("Eg v T")
plt.legend(frameon=False)
plt.ylabel("Eg (eV)")
plt.xlabel("T (K)")
plt.savefig("ps2_q2c.png")
```
